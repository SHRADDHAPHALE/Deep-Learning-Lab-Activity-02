**QUESTION 1**

**Demonstrate the activation function using python code**

```
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```
a = np.arange(-5, 5, 0.2)
```

Sigmoidal activation function

```
def sigmoid(z):
    return 1/(1+np.exp(-z))

b = sigmoid(a) 
plt.plot(a,b)
plt.title("Sigmoid")
plt.grid()
```

Tanh activation function

```
def tanh(z):
    return (np.exp(z) - np.exp(-z))/(np.exp(z)+np.exp(-z))

b = tanh(a)
plt.plot(a,b)
plt.title("Tanh")
plt.grid()
```

ReLU activation function

```
def ReLU(z):
    z_zero = np.array([])
    z_max = np.array([])
    z_result = np.array([])
    for i in z:
        if i > 0:
            z_max = np.append(z_max,i)
        else:
            i = 0
            z_zero = np.append(z_zero,i)
    z_result = np.concatenate((z_zero,z_max))
    return z_result

b = ReLU(a)
plt.plot(a,b)
plt.title("ReLU")
plt.grid()
```

Leaky ReLU activation function

```
def leakyReLU(z,alpha = 0.2):
    z_zero = np.array([])
    z_max = np.array([])
    z_result = np.array([])
    for i in z:
        if i > 0:
            z_max = np.append(z_max,i)
        else:
            i = alpha * i 
            z_zero = np.append(z_zero,i)
    z_result = np.concatenate((z_zero,z_max))
    return z_result

b = leakyReLU(a)
plt.plot(a,b)
plt.title("Leaky ReLU")
plt.grid()
```

Exponential ReLU activation function

```
def elu(z, alpha):
    z_result = []
    for i in z:
        if i >= 0:
            z_result.append(i)
        else:
            z_result.append(alpha * (np.exp(i)-1))
    return z_result

b = elu(a, 0.2)

plt.plot(a, b)
plt.title('Exponential ReLU')
plt.grid()
```
